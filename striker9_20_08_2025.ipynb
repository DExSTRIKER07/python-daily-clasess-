```python
# Compile time / Method Overloading Polymorphism
class BCA_1St_Semester :
  def student (self, a,b=0,c=0,d=0):
      return a+b+c+d

BCA = BCA_1St_Semester()
print(BCA.student(5))
print(BCA.student(10,7,8))
print(BCA.student(-10,4,9,-5))
print(BCA.student(99,0,4,-12))
```

```python
#Run time / Method overriding
class Animal :
    def sound(self):
      print("Animal makes a sound")

class Dog :
    def sound(self):
      print("Bow Bow Bow")

class Cat :
    def sound(self):
      print("Mwoe Meow Meow")

class Snake :
    def sound(self):
      print("bussssssss")

class Cow :
    def sound(self):
      print("Ambaaaaaaaa")

obj1 = Animal()
obj2 = Dog()
obj3 = Cat()
obj4 = Snake()
obj5 = Cow()

obj2.sound()
obj1.sound()
obj5.sound()
obj4.sound()
obj3.sound()
```

```python
# Operator overloading polymorphism
class Book :
  def __init__(self,page) :
    self.page = page
  def __add__(self,other) :
    return self.page + other.page

obj1 = Book(200)
obj2 = Book(300)
print("Total pages is ", obj1 + obj2)
```

```python
# Abstraction example
from abc import ABC, abstractmethod

class Game(ABC):
    def Acting(self):
        print("I'm doing work")

    # Normal function / method
    def BGMI(self):
        pass

    #Abstract method / function
    @abstractmethod
    def ludo_king(self):
        pass

class Temple(Game):
    # BGMI function implemented by sub class
    def BGMI(self):
        print("I'm got diamond")

    # Normal game
    def Chess(self):
        print("I'm dist player")

    # Implementing the abstract method
    def ludo_king(self):
        print("I'win the game")


obj = Temple()
obj.BGMI()
obj.ludo_king()
obj.Chess()
obj.Acting()
```
